```python
import tensorflow as tf
import numpy as np
import cv2
import pickle
```

```python
face = "C:/Users/shahe/AI PROJECTS/Instagram Custom Filters/haarcascade_frontalface_default.xml"
face_cascade = cv2.CascadeClassifier(face)
```

```python
def get_model():
    backend = tf.keras.applications.EfficientNetB2(
        input_shape=(96,96,3),
        include_top = False,
        weights = None
    )
    model = tf.keras.Sequential([
        backend,
        tf.keras.layers.GlobalAveragePooling2D(),
        tf.keras.layers.Dropout(0.3),
        tf.keras.layers.Dense(128, activation='relu'),
        tf.keras.layers.Dense(7, activation='softmax')
    ])
    return model

model = get_model()
model.load_weights("C:/Users/shahe/Desktop/face expression/FacialExpressionModel.h5")
```

```python
def load_object(name):
    pickle_obj = open(f"{name}.pck","rb")
    obj = pickle.load(pickle_obj)
    return obj

Le = load_object("C:/Users/shahe/Desktop/face expression/LabelEncoder")
```

```python
video_capture = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30)
    )

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        image = cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        image = tf.convert_to_tensor(image)
        image = tf.image.resize(image , [96 , 96] , method="bilinear")
        image = tf.expand_dims(image , 0)
        pred = model.predict(image)
        pred = np.argmax(pred , axis = 1)
        Le.inverse_transform(pred)[0]
        print(pred)

    # Display the resulting frame
    cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
               
video_capture.release()
cv2.destroyAllWindows()
```
